```python
from tqdm import tqdm

# First, ensure your date index is properly formatted
df_jour['date'] = pd.to_datetime(df_jour['date'])

# Select only numeric columns for interpolation
numeric_columns = df_jour.select_dtypes(include=['float64', 'int64']).columns

# Create a copy of the dataframe
df_interpolated = df_jour.copy()

# Set the datetime index
df_interpolated.set_index('date', inplace=True)

# Sort the index (important for time-based interpolation)
df_interpolated.sort_index(inplace=True)

# Interpolate only numeric columns
for col in tqdm(numeric_columns, desc="Interpolating columns"):
    df_interpolated[col] = df_interpolated[col].interpolate(
        method='nearest' , limit_direction='both', limit_area='inside'
    )
print(df_interpolated.isna().sum())
# For any remaining NaN values after interpolation, forward fill with a limit
df_interpolated[numeric_columns] = df_interpolated[numeric_columns].fillna(
    method='ffill',
    limit=6  # Forward fill up to 6 time steps
)
```

# Enrichissement des données (Data Enrichment)

```python
def enrich_with_season(df):
    """
    Enrichir un DataFrame avec une colonne 'season' basée sur le mois de la date
    et appliquer un encodage one-hot pour les saisons. Les colonnes encodées seront converties en entiers (0 ou 1).
    """
    def assign_season(month):
        """
        Assigner une saison basée sur le mois.
        """
        seasons = {
            (12, 1, 2): 'winter',   # Hiver
            (3, 4, 5): 'spring',    # Printemps
            (6, 7, 8): 'summer',    # Été
            (9, 10, 11): 'autumn'   # Automne
        }
        for months, season in seasons.items():
            if month in months:
                return season

    # Extraire le mois à partir de la colonne 'Date' ou de l'index
    if 'Date' in df.columns:
        df['Month'] = pd.to_datetime(df['Date']).dt.month
    else:
        df['Month'] = df.index.month

    # Appliquer la logique des saisons
    df['season'] = df['Month'].apply(assign_season)

    # Encodage one-hot pour la colonne 'season'
    season_encoded = pd.get_dummies(df['season'], prefix='season')

    # Convertir les colonnes encodées en entiers
    season_encoded = season_encoded.astype(int)

    # Ajouter les colonnes encodées et la colonne 'Month' au DataFrame
    df = pd.concat([df, season_encoded], axis=1)

    # Supprimer la colonne 'season' mais conserver 'Month'
    df.drop(columns=['season'], inplace=True)
    
    return df
```

```python
# Enrichir chaque DataFrame
df_train = enrich_with_season(df_train)
df_validation = enrich_with_season(df_validaton)
df_test = enrich_with_season(df_test)

# Afficher l'ensemble enrichi pour vérification
df_train
```

```python
def enrich_with_elapsed_time(df, sunrise_col):
    df = df.copy()
    
    # Réinitialiser l'index pour ne pas laisser la date comme index
    df.reset_index(inplace=True)
    
    # Calculer la durée écoulée entre la colonne 'date' (index réinitialisé) et la colonne de sunrise
    df['elapsed_time'] = (
        df['date'].dt.tz_localize(None) - df[sunrise_col].dt.tz_localize(None)
    )
    
    # Convertir la durée en secondes
    df['elapsed_time_seconds'] = df['elapsed_time'].dt.total_seconds()
    
    # Supprimer la colonne intermédiaire 'elapsed_time'
    df.drop(columns=['elapsed_time'], inplace=True)
    
    return df
```

```python
# Enrichir les ensembles de données
df_train = enrich_with_elapsed_time(df_train, 'Sunrise')
df_validation = enrich_with_elapsed_time(df_validation, 'Sunrise')
df_test = enrich_with_elapsed_time(df_test, 'Sunrise')

# Vérification des résultats
df_train.head()
df_validation.head()
df_test.head()
```

# Normalisation des données (Data Normalization) :

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import numpy as np

def normalize_data(df_train, df_validation, df_test, transformations=None):
    # Créer des copies des DataFrames pour les ensembles de données
    normalized_train = df_train.copy()
    normalized_validation = df_validation.copy()
    normalized_test = df_test.copy()

    # Si aucune transformation spécifique n'est donnée, appliquer Min-Max Scaling par défaut
    if transformations is None:
        transformations = {col: ['minmax'] for col in df_train.select_dtypes(include=['float64', 'int64']).columns}

    # Dictionnaires de scalers
    min_max_scaler = MinMaxScaler()
    standard_scaler = StandardScaler()

    # Appliquer les transformations par colonne
    for col, trans in transformations.items():
        # Appliquer Min-Max Scaling
        if 'minmax' in trans:
            # Normaliser les données avec Min-Max
            normalized_train[col] = min_max_scaler.fit_transform(normalized_train[[col]])
            normalized_validation[col] = min_max_scaler.transform(normalized_validation[[col]])
            normalized_test[col] = min_max_scaler.transform(normalized_test[[col]])
        
        # Appliquer Standard Scaling
        if 'standard' in trans:
            # Standardiser les données
            normalized_train[col] = standard_scaler.fit_transform(normalized_train[[col]])
            normalized_validation[col] = standard_scaler.transform(normalized_validation[[col]])
            normalized_test[col] = standard_scaler.transform(normalized_test[[col]])

        # Appliquer Log Transformation + Min-Max Scaling
        if 'logminmax' in trans:
            # Appliquer la transformation logarithmique
            normalized_train[col] = np.log1p(normalized_train[col])
            normalized_train[col] = min_max_scaler.fit_transform(normalized_train[[col]])
            normalized_validation[col] = np.log1p(normalized_validation[col])
            normalized_validation[col] = min_max_scaler.transform(normalized_validation[[col]])
            normalized_test[col] = np.log1p(normalized_test[col])
            normalized_test[col] = min_max_scaler.transform(normalized_test[[col]])

    # Retourner les DataFrames normalisés
    return normalized_train, normalized_validation, normalized_test

# Exemple d'application avec l'ajout de 'elapsed_time_seconds'
transformations = {
    'Active_Power': ['minmax'],
    'Weather_Temperature_Celsius_y': ['standard'],
    'Weather_Relative_Humidity_y': ['minmax'],
    'Global_Horizontal_Radiation_y': ['logminmax'],  # Applique log + MinMax
    'elapsed_time_seconds': ['minmax'],  # Applique Min-Max Scaling à elapsed_time_seconds
    'maintenance_startm': ['minmax'],
    'maintenance_starth': ['minmax'],
    'maintenance_endm': ['minmax'],
    'maintenance_endh': ['minmax']
}
df_train_normalized, df_validation_normalized, df_test_normalized = normalize_data(df_train, df_validation, df_test, transformations)

# Affichage des résultats pour vérifier la normalisation
df_test_normalized.head()
```

# Transformation des données (Data Reshape) :

```python
import numpy as np

# Définir la longueur de la séquence
sequence_length = 10  # Exemple : 10 pas de temps
def create_sequences(data):
    # Initialisation des listes pour stocker les séquences
    X_sequences = []  # Caractéristiques (features)
    Y_sequences = []  # Cibles (target)

    # Créer des séquences glissantes
    for i in range(len(data) - sequence_length + 1):
        # Extraire les caractéristiques (tout sauf 'Active_Power')
        X_seq = data[i:i + sequence_length, 1:]  # Exclure la colonne 'Active_Power' (colonne 0)
        
        # Extraire la variable cible 'Active_Power'
        Y_seq = data[i + sequence_length - 1, 0]  # La dernière valeur de 'Active_Power' dans la séquence
        
        X_sequences.append(X_seq)
        Y_sequences.append(Y_seq)

    # Convertir les listes en arrays numpy
    X_sequences = np.array(X_sequences)
    Y_sequences = np.array(Y_sequences)
    
    return X_sequences, Y_sequences

# Assurez-vous que les jeux de données sont préparés
train_data = df_train_normalized.values  # Normalized training data
val_data = df_validation_normalized.values  # Normalized validation data
test_data = df_test_normalized.values  # Normalized test data
```

```python
# Créer les séquences pour chaque jeu de données
X_train, Y_train = create_sequences(train_data)
X_val, Y_val = create_sequences(val_data)
X_test, Y_test = create_sequences(test_data)
```

```python
# Vérifiez la forme des données reshaped
print("Shape des caractéristiques (X_train) reshaped:", X_train.shape)  # (samples, sequence_length, features)
print("Shape des cibles (Y_train):", Y_train.shape)  # (samples,)

print("Shape des caractéristiques (X_val) reshaped:", X_val.shape)  # (samples, sequence_length, features)
print("Shape des cibles (Y_val):", Y_val.shape)  # (samples,)

print("Shape des caractéristiques (X_test) reshaped:", X_test.shape)  # (samples, sequence_length, features)
print("Shape des cibles (Y_test):", Y_test.shape)  # (samples,)
```

# Entrainer un modèle LSTM 

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Définir le modèle
model = Sequential()
model.add(LSTM(64, input_shape=(10, 14))) # 10 est la sequence_length, 9 est le nombre de features
model.add(Dense(1)) # Pour une régression, une seule unité de sortie est nécessaire

# Compiler le modèle
model.compile(optimizer='adam', loss='mse') # mse pour la régression

# Entraîner le modèle
model.fit(X_train, Y_train, epochs=10, validation_data=(X_val, Y_val)) # Ajustez le nombre d'époques selon vos besoins

# Évaluer le modèle
loss = model.evaluate(X_test, Y_test)
print(f"Loss: {loss}")

# Faire des prédictions
predictions = model.predict(X_test)
```
